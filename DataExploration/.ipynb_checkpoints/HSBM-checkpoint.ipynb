```python
import os
import pylab as plt
import json
%matplotlib inline

from sbmtm import sbmtm
import graph_tool.all as gt
```

# Create corpus

```python
tweets_file = '2019_april_may_uber.txt'
tweets = []
with open(tweets_file, 'r') as f:
    for line in f:
        tweet = json.loads(line)
        tweets.append(tweet)        

tweets_texts = [t['tweet'].split() for t in tweets]
tweets_titles = [t['tweet'] for t in tweets]

dates = [t['date'] for t in tweets]

print(dates)
```

```python
print(tweets_titles[2])
print(tweets_texts[2])
print(tweets[2])
```

```python
model = sbmtm()

model.make_graph(tweets_texts[:10_000], documents=tweets_titles[:10_000])
gt.seed_rng(32)
```

```python
model.fit()
```

# Plotting the result

```python
model.plot(nedges=1000)
```
